En este notebook trabajaremos con $\textit{Mathplotlib}$. (https://matplotlib.org/)
Matplotlib es una libreria muy útil y versatil para crear gráficas en Python. Con esta libreria, entre otras cosas, podemos:
- Crear gráficos.
- Hacer figuras interactivas que se puedan ampliar, desplazar, actualizar.

En resumen es muy útil.

```python
#import matplotlib #Aquí le decimos que importe matplotlib, pero genenralmente se usa como:
import matplotlib.pyplot as plt #es decir, le pedimos que importe la librería y que le ponga "un apodo", un nombre más corto y fácuil de usar. Es convención pedir que importen matplotlib como plt.

plt.plot([1, 2, 3, 4]) #Aquí le decimos al notebook qué ha de graficar. En este caso es un array con cuatro valores, y lo llamamos directamente para graficarlo.
plt.xlabel('Eje x ') #Aquí le pedimos que le ponga una etiqueta o label al eje x.
plt.ylabel('Eje y') #Aquí le pedimos que ponga le ponga una leyenda al eje y.
plt.title('Números') #Aquí le pedimos que agregue un nombre o título para la gráfica.
plt.show()
```

```python
# Aquí generamos algunos datos para graficar y jugar con ellos un poco:
x = np.linspace(0, 2*np.pi, 400) #Igual regresa datos espaciados uniformemente dentro de un intervalo.
y = np.sin(x**2)
```

```python
# Crea uan figura y una subfigura o subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Una subgráfica')
```

```python
# Crea dos subplots
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True) #1 subfigura en el eje x y dos en el eje y, que comparten el eje y.
ax1.plot(x, y)
ax1.set_title('Título de la subgráfica 1.')
ax2.scatter(x, y)
ax2.set_title('Título de la subgráfica 2.')
```

```python
# Gráficas en coordenadas polares.
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection="polar"))
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y, marker = '*', s = 1, c = y )
axs[0,1].plot(y,x)
```

```python
# Estas 4 subgráficas comparten el eje x, por columna.
#Aquí le pedimos que cree 2 subfijuras en el el eje x y 2 en el y, pero podríamos pedirle que fueran 3 en x; 2 en y (3,2) o 2 en x y 5 en y (2,%).
#Pero ten cuidado porque también deberás modificar las dimensiones de las leyendas y/o la gráfica o no será clara la información que presentes.
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('(1)') #Las subgráficas (1) y (3) comparaten eje x.
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('(2)')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('(3)') #Las subgráficas (3) y (4) comparten eje y.
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('(4))')

for ax in axs.flat:
    ax.set(xlabel='Eje x', ylabel='Eje y')

# Esta partecita del código es la que esconde las etiquetas de los ejes x y y para las gráficas superiores y a la derecha, respectivamente.
for ax in axs.flat:
    ax.label_outer()
```

#Grafica en 3d

```python
#3dplot

import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')

t=np.linspace(-2,2,100)
theta=np.linspace(-2*np.pi,2*np.pi,100)
r=t**2+1
x=r*np.sin(theta)
y=r*np.cos(theta)

ax.plot(x,y,t)
plt.show()
```

```python
angulo= np.linspace(0,2*np.pi,30)

theta=np.meshgrid(angulo) #angulo axial
phi=np.meshgrid(angulo) #anglo azimutal
r,R=0.25,1.

def f1(theta,phi):#x
   return(R+r*np.cos(phi)*np.cos(theta))

def f2(theta,phi):

  return((R+r*np.cos(phi))*np.sin(theta))

def f3(phi):
  return (r*np.sin(phi))

fig =plt.figure()
ax= fig.add_subplot(111,projection='3d')
ax.set_zlim3d(-2,2)
ax.plot_surface(f1(theta,phi),f2(theta,phi),f3(phi))
plt.show()
```
