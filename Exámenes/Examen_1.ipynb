##Ejercicio 1. Diseño de un telescopio refractor


Un fabricante de telescopios está diseñando un telescopio refractor utilizando lentes convergentes. El telescopio constará de dos lentes: una lente objetivo de distancia focal $ f_o $ y una lente ocular de distancia focal $ f_e $. El objetivo del telescopio es producir una imagen ampliada y nítida de objetos lejanos.

Para calcular la magnificación $ M $ del telescopio y la distancia focal del mismo se tiene la siguiente
fórmula:

$$ M = \frac{f_e}{f_o} $$



Mientras que la distancia focal del sistema, denotada como $ f_s $, se expresa en terminos del lente objetivo y del lente ocular:

$$ \frac{1}{f_s} = \frac{1}{f_o} + \frac{1}{f_e} $$

$\textbf{a)}$ El actual diseño del telescopio tiene los siguientes valores, para la distancia focal de la lente objetivo $ f_o $ = 0.1 metros, y para la distancia focal de la lente ocular $ f_e $ = 0.02 metros. Calcula la magnificación del telescopio y la distancia focal del sistema.


$\textbf{b)}$ Como parte del trabajo el fabricante está probando diferentes distancias focales del sistema para obtener mejores magnificaciones. Realiza una gráfica de la magnificación en función de la distancia focal del sistema. Si la distancia focal del sistema toma valores entre  0.01 - 2 metros, que tu gráfica tenga almenos 100 puntos.

La magnificación del telescopio $M$ en función de la distancia focal del sistema $ f_s $ se puede calcular como:

$$ M = \frac{f_e}{f_s} -1$$

Toma $ f_e = 0.02 m$ y calcula M para diferentes valores de $ f_s $

```python
#inciso a)
def magnificacion(fo,fe):
  M=fe/fo
  return M
print("la magnificación es:", magnificacion(0.1,0.02),"metros")

def distancia_focal_sistema (fo,fe):
  fs= (fo*fe)/(fo+fe )
  return fs
print("La distancia focal del sistema es:", distancia_focal_sistema(0.1,0.02),"metros")
```

## Ejercicio 2. Circuito RC en un circuito de audio

Se tiene un circuito RC en serie conectado a una fuente de corriente alterna con voltaje $ V_0 = 10 $ V, la resistencia $ R $ del circuito es de 100 Ω y la capacitancia $ C $ es de 1 μF.


La impedancia $ Z $ del circuito se calcula como:

$$ Z = R + \frac{1j}{\omega C} $$

Donde la j significa número complejo, es decir debes de tomar el 1 complejo y dividirlo entre el producto de $ \omega C$.

La amplitud de la corriente en el circuito se calcula como:

$$ I_{\text{amplitud}} = \frac{V_0}{|Z|} $$

donde:

- $ V $ es el voltaje en el circuito (en volts),
- $ I $ es la corriente en el circuito (en amperios),
- $ R $ es la resistencia en el circuito (en ohma),
- $ C $ es la capacitancia en el circuito (en faradios), y
- $ \omega = 2\pi f $ es la frecuencia angular.


$\textbf{a)}$ Averigua la relación entre la amplitud de la corriente y la frecuencia y escribe una función que la calcule. Supón que la frecuencia $f$ varia entre 1 Hz y 10 kHz (kilohertz) y realiza una gráfica para visualizar esta relación. Procura tener al menos 500 valores de frecuencia en el rango mencionado. Para cada uno de ellos correspondera un valor de amplitud.

Hint: en la misma función calcula los valores de la frecuencia angular y la impedancia. Presta mucha atencion al numero complejo.

$\textbf{b)}$ Discute **con tus palabras** el significado de la gráfica, trata de dar un sentido físico al comportamiento de la amplitud. Para visualizar mejor la gráfica usa escala logarítmica en el eje x, utiliza la instrucción $\textit{plt.xscale('log')}$ para modificar la escala. Esta instruccion la debes agregar despues de decirle a python los valores que quieres graficar.

```python
#Tu código va aquí.
from math import *
import matplotlib.pyplot as plt
V0=10 #V
R=100 #Ohms
C= 0.000001 #mf
frecuencias=np.linspace(1,10000,500)
I=[]

def relacion(V0,R,C,f):
  omega=2*np.pi*f
  Z= R+ ((0+1j)/(omega*C))
  Iamp= V0/ abs(Z)
  return Iamp


for i in frecuencias:
  I.append(relacion(V0,R,C,i))

plt.plot(frecuencias, I,"c")
plt.xlabel('Frecuencias (kHz)')
plt.ylabel('Amplitud')
plt.title('Amplitud de corriente vs frecuencias')
plt.xscale('log')
plt.show()


```

# b) Interpretación física

La gráfica expresa una relación entre la amplitud de corriente y la frecuencia(kHz) la cual se comporta de manera exponencial, al menos de forma local. Para valores superiores a  $10^2$ $kHz$ se observa un incremento abrupto en la gráfica, si consideramos valores superiores a $10^3$ $kHz$ podemos ver que se alcanza un máximo local en la función muy cercano a 0.1, ahora bien, si el valor para la frecuencia es menor que $10^2 kHz$ la amplitud apenas y sufre cambios significativos.

Es decir que si quisieramos observar cambios en la amplitud de corriente en el laboratorio, lo que se tendría que hacer es aumentar la frecuencia a valores mayores a $10^2$ $kHz$ para observar tales cambios.
